```python
import pandas as pd
import csv
```

```python
filepath_NH_2015 = 'New Hampshire/NH_2015_nibrs_arrestee.csv'
filepath_NH_2016 = 'New Hampshire/NH_2016_nibrs_arrestee.csv'
filepath_NH_2017 = 'New Hampshire/NH_2017_NIBRS_ARRESTEE.csv'
filepath_NH_2018 = 'New Hampshire/NH_2018_NIBRS_ARRESTEE.csv'
filepath_NH_2019 = 'New Hampshire/NH_2019_NIBRS_ARRESTEE.csv'
```

```python
NH_2015_df = pd.read_csv(filepath_NH_2015)
NH_2016_df = pd.read_csv(filepath_NH_2016)
NH_2017_df = pd.read_csv(filepath_NH_2017)
NH_2018_df = pd.read_csv(filepath_NH_2018)
NH_2019_df = pd.read_csv(filepath_NH_2019)
```

```python
NH_2017_df_2 = NH_2017_df.rename(columns={'ARREST_DATE':'arrest_date','OFFENSE_TYPE_ID':'offense_type_id','SEX_CODE':'sex_code'})
NH_2018_df_2 = NH_2018_df.rename(columns={'ARREST_DATE':'arrest_date','OFFENSE_TYPE_ID':'offense_type_id','SEX_CODE':'sex_code'})
NH_2019_df_2 = NH_2019_df.rename(columns={'ARREST_DATE':'arrest_date','OFFENSE_TYPE_ID':'offense_type_id','SEX_CODE':'sex_code'})
NH_2016_df.head()
```

```python
merge_1 = pd.merge(
    NH_2015_df,NH_2016_df, on = ["arrest_date",'offense_type_id'], how = "outer")
merge_2 = pd.merge(
    merge_1,NH_2017_df_2, on = ["arrest_date",'offense_type_id'], how = "outer")
merge_3 = pd.merge(
    merge_2,NH_2018_df_2, on = ["arrest_date",'offense_type_id'], how = "outer")
NH_DF_Dirty = pd.merge(
    merge_3,NH_2019_df_2, on = ["arrest_date",'offense_type_id'], how = "outer")
```

# date separate for year and month
# no offense type
# Add state column
# filter by 35

```python
NH_DF_Dirty['State'] = 'NH'
month = ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL',
        'AUG', 'SEP', 'OCT', 'NOV', 'DEC']
date = NH_DF_Dirty['arrest_date'].str.split('-')
month = []
for number in date:
    if number[1] == 'JAN':
        month.append(1)
    if number[1] == 'FEB':
        month.append(2)
    if number[1] == 'MAR':
        month.append(3)
    if number[1] == 'APR':
        month.append(4)
    if number[1] == 'MAY':
        month.append(5)
    if number[1] == 'JUN':
        month.append(6)
    if number[1] == 'JUL':
        month.append(7)
    if number[1] == 'AUG':
        month.append(8)
    if number[1] == 'SEP':
        month.append(9)
    if number[1] == 'OCT':
        month.append(10)
    if number[1] == 'NOV':
        month.append(11)
    if number[1] == 'DEC':
        month.append(12)
```

```python
NH_DF_Clean = NH_DF_Dirty.loc[NH_DF_Dirty['offense_type_id'] == 35,['arrest_date','State','sex_code']]
```
